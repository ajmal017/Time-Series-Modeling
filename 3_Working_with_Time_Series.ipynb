#### From text to date

```
df.date = pd.to_datetime(df.date, dayfirst=True)
df.head()
```

```
df.date.describe()
```

#### Setting the index

```
df.set_index("date", inplace=True)
```

```
df.head()
```

```
df.date.describe()

# it throws an error because once date becomes an index, it is no longer a seperate attribute in the dataframe.
```

#### Setting the frequency

```
# When we wish to define the frequency by day
df = df.asfreq('d')
df.head()
```

```
# What if we wish to not include the weekends because the stock markets are not operative?

df = df.asfreq('b')
df.head()
```

#### Handling missing values

```
df.isna()
```

```
df.isna().sum()

# We see there are 8 missing values for each stock market.
# Setting the frequency to business days thus generated these 8 missing values.
```

```
# Filling data
# 1. Front filling
# 2. Back filling
# 3. Assign the same value

# Front fill the spx
df.spx = df.spx.fillna(method = "ffill")
```

```
df.isna().sum()
```

```
# Back fill the ftse
df.ftse = df.ftse.fillna(method = "bfill")

# Average fill the dax and nikkei
df.dax = df.dax.fillna(value = df.dax.mean())
df.nikkei = df.nikkei.fillna(value = df.nikkei.mean())
```

```
df.isna().sum()
```

#### Simplifying the dataset

```
df['market_value'] = df.spx
```

```
df.describe()
```

```
del df['spx'], df['ftse'], df['dax'], df['nikkei']
```

```
df.describe()
```

#### Splitting the data

In timeseries we must never randomize the training and test set. It must be in the same sequence.

We will do a 80:20 split in this case

```
size = int(len(df)*0.8)
```

```
df_train = df.iloc[:size]
df_test = df.iloc[size:]
```

##### Verify the data split

```
df_train.tail()
```

```
df_test.head()

# We see the sequence has been properly maintained. The couple of days missing are the weekends.
```

#### Working with white noise

```
# Importing more packages relevant to our ML exercise on time series data

import statsmodels.graphics.tsaplots as sgt
import statsmodels.tsa.stattools as sts
from statsmodels.tsa.seasonal import seasonal_decompose
import seaborn as sns
sns.set()
```
