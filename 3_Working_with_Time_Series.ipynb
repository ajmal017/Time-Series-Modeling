#### Stationarity

We check for Dickey - Fuller test

```
sts.adfuller(df.market_value)

# First value is t-stat, and second is p-value

# Here, t-stat is greater than all three levels of confidence and thus we can confirm the stationarity

# 3rd line represents number of lags used in the regression while calculating the t-stat. Here we have 10. Therefore, there is some auto-corelation

# 4th line shows number of observations used for analysis

# Last line shows maximized information criteria given some auto-corelation. Lower the value, better we chances we have for prediction
```

```
sts.adfuller(df.wn)
```

#### Seasonality

1. Trend -- Tells us about the pattern in the data
2. Seasonal -- Cyclical effects
3. Residual - Errors in prediction

Naive Decomposition:

Additive: Observed = trend + seasonal + residual

Multiplicative: Observed = trend * seasonal * residual

```
# Additive

s_dec_additive = seasonal_decompose(df.market_value, model = "additive")
s_dec_additive.plot()
plt.show()
```

```
# Multiplicative

s_dec_multiplicative = seasonal_decompose(df.market_value, model = "multiplicative")
s_dec_multiplicative.plot()
plt.show()
```

#### ACF

Autocorrelation is about the correlation between a sequence and itself.

Lags is looking at the past data (in days, months or years) and its correlation to the present date.

```
sgt.plot_acf(df.market_value, lags = 40, zero = False)
plt.title("ACF S&P", size = 24)
plt.show()

# The blue shadow region tells us the significance.
# The greater the distance as we move from left to right, tells us it is more unlikely that the autocorrelation persists.
```

```
sgt.plot_acf(df.wn, lags = 40, zero = False)
plt.title("ACF WN", size = 24)
plt.show()

# No autocorrelation in the white noise as seen from the significant area.
```

#### PACF

Partial Autocorrelation

ACF and PACF for first lag is always the same.

ACF shows both direct and indirect effect. PACF only shows the direct effect.

```
sgt.plot_pacf(df.market_value, lags = 40, zero = False, method = ('ols'))
plt.title("PACF S&P", size = 24)
plt.show()

# It cancels out all additional channels a previous period value affects the present one.
```

```
sgt.plot_pacf(df.wn, lags = 40, zero = False, method = ('ols'))
plt.title("PACF WN", size = 24)
plt.show()

# No autocorrelation in the white noise as seen from the significant area.
```
